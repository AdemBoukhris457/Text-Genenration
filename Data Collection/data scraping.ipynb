```python
def get_year_details(url):
    
    pros = []
    cons = []
    description1 = []
    description2 = []
    description3 = []
    description4 = []
    description5 = []
    description6 = []
    full_block = []

    all_pros_details = []
    all_cons_details = []
    all_descriptions_details = []
    
    all_descs = []
    pros_1 = []
    cons_1 = []
    pros_details = []
    cons_details = []
    
    
    r = s.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    #block = soup.find_all("div", {"name":"subnav-reviews"})
    pros = soup.find("ul", {"aria-labelledby":"view_pros", "data-tracking-parent":"view_pros"})
    if pros is not None:
        pros_1 = pros.find_all("span")
        for i in range(len(pros_1)):
            pros_details.append(pros_1[i].text)
    else:
        pros_details = "empty"
    #print(pros_details)
    
    cons = soup.find("ul", {"aria-labelledby":"view_cons", "data-tracking-parent":"view_cons"})
    if cons is not None:
        cons_1 = cons.find_all("span")
        for i in range(len(cons_1)):
            cons_details.append(cons_1[i].text)
    else:
        cons_details = "empty"
    #print(cons_details)
    
    description1 = soup.find_all("div", {"class":"content-fragment review-section-fragment size-16"}) #p
    description2 = soup.find_all("div", {"class":"content-fragment review-section-fragment size-16 mb-1_5"}) #p
    description3 = soup.find_all("div", {"class":"content-fragment review-section-fragment text-gray-darker size-16 pl-1_5"})#div
    description4 = soup.find_all("div", {"class":"content-fragment editorial-review-section-fragment size-16"}) #
    
    #print(description1)
    for i in range(len(description1)):
        #print(description1[i].text)
        all_descs.append(description1[i].text)
        

    #print(description2)
    for i in range(len(description2)):
        #print(description2[i].text)
        all_descs.append(description2[i].text)
        
    
    #print(description3)
    for i in range(len(description3)):
        #print(description3[i].text)
        all_descs.append(description3[i].text)
        
    
    #print(description4)
    for i in range(len(description4)):
        #print(description4[i].text)
        all_descs.append(description4[i].text)
    
    return pros_details, cons_details, all_descs
```

```python
all_pros_details, all_cons_details, all_descriptions_details= get_year_details(years_hrefs[0])
```
